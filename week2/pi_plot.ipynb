```python
import numpy as np
import math
import random
import matplotlib.pyplot as plt

def pi_estimate(N):
    x = np.random.uniform(-1, 1, int(N))
    y = np.random.uniform(-1, 1, int(N))
    distances = x**2 + y**2
    inside_mask = distances <= 1
    count = np.sum(inside_mask)
    return (4 * count / N)

N_values = np.logspace(2, 6, num=200, dtype=int)
pi_estimates = [pi_estimate(n) for n in N_values]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(N_values, pi_estimates, marker='o', linestyle='-', label='Estimated ')
plt.xscale('log')
plt.xlabel('Number of Points (N)')
plt.ylabel('Estimated Value of ')
plt.title('Estimation of  using Monte Carlo Method')
plt.legend()
plt.grid(True)
plt.axhline(y=math.pi, color='r', linestyle='--', label='Actual ')

plt.show()
```

```python
percent_errors = [abs((est - math.pi) / math.pi) * 100 for est in pi_estimates]
plt.plot(N_values, percent_errors, marker='o', linestyle='-', label='Percent Error')
plt.xscale('log')
plt.xlabel('Number of Points (N)')
plt.ylabel('Percent Error (%)')
plt.legend()
plt.grid(True)

plt.show()
```
