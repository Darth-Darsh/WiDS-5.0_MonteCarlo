```python
import numpy as np
import random
import math
```

```python
N = 100_000
x = np.linspace(0, np.pi, N)
y = np.sin(x)
area = 0
for i in range(N - 1):
    area += (y[i]) * (x[i + 1] - x[i]) 

print("Estimated area using Reimann Sums:", area)
```

```python
def monte_carlo(N):
    x = np.random.uniform(0, np.pi, N)
    y = np.random.uniform(0, 1, N)
    inside_mask = y <= np.sin(x)
    count = np.sum(inside_mask)
    return (count / N) * (np.pi)

print("Estimated area using Monte Carlo:", monte_carlo(N))
```

```python
true_value = 2.0

reimann_error = abs(true_value - area)
monte_carlo_error = abs(true_value - monte_carlo(N))

print("Reimann Sums Error:", reimann_error)
print("Monte Carlo Error:", monte_carlo_error)
```
