```python
import numpy as np
import math
import random
```

```python
n = 10

dx = 2 / 4
points = np.linspace(-1 + dx/2, 1 - dx/2, 4)

grid = np.meshgrid(*[points]*n)

coords = np.stack(grid, axis=0)


r2 = np.sum(coords**2, axis=0)

inside = r2 <= 1

volume = np.sum(inside) * (dx ** n)

print("Estimated volume of unit sphere in 10D:", volume)
```

```python
N = 1_000_000
x = np.zeros((11, 1_000_000))
def monte_carlo(N):
    sum = 0
    for i in range(1, 11):
        x[i] = np.random.uniform(-1, 1, N)
        sum += x[i]**2
    inside_mask = sum <= 1
    count = np.sum(inside_mask)
    return (count / N) * (2**10)

monte_value = monte_carlo(N)

print("Estimated volume using Monte Carlo:", monte_value)   
```

```python
true_value = np.pi**5 / 120
monte_carlo_error = abs(true_value - monte_value)
grid_error = abs(true_value - volume)

print("Grid-based Error:", grid_error)

print("Monte Carlo Error:", monte_carlo_error)
```
